```r
# ---------------------------------------------
# STEP 1: Install and load required libraries
# ---------------------------------------------
install.packages("tidyverse")
install.packages("skimr")
install.packages("janitor")

library(tidyverse) # For data wrangling and visualization
library(skimr) # For quick data overview
library(janitor)  # For cleaning column names
```

```r
# ---------------------------------------------
# STEP 2: Load data
# ---------------------------------------------

stroke <- read_csv("healthcare-dataset-stroke-data.csv")

# Preview first rows
head(stroke)
```

```r
# ---------------------------------------------
# STEP 3: Clean column names
# ---------------------------------------------

stroke <- stroke %>% clean_names()

head(stroke)
```

```r
# Quick summary of dataset
summary(stroke)
```

```r
# Structure of dataset (column types)
str(stroke)
```

```r
# Basic info
glimpse(stroke)
```

```r
# ---------------------------------------------
# STEP 4: Handle missing values
# ---------------------------------------------

# Count missing value per column
colSums(is.na(stroke))
```

```r
# -------------------------------------------------------------
# Problem:
# colSums(is.na()) shows 0 missing values, but the dataset
# actually contains "N/A", "", and "Unknown" as *text strings*,
# which R does NOT treat as real NA.
# This step identifies and converts them to TRUE NA.
# -------------------------------------------------------------
```

```r
# -------------------------------
# 4.1 CHECK "FAKE" MISSING VALUES
# -------------------------------
# We count how many values are "N/A", empty string "",
# or "Unknown" (especially in smoking_status).
# These are not real NA, but they represent missing data.

sapply(stroke, function(x) sum(x %in% c("N/A", "", "Unknown")))
```

```r
# ---------------------------------------------
# 4.2 CONVERT "N/A", "", "Unknown" → TRUE NA
# ---------------------------------------------
# We convert all the fake missing values into TRUE NA.
# na_if(x, "value") replaces the specified value with NA.
# We clean bmi and smoking_status mostly because they contain text gaps.

library(dplyr)

stroke_clean <- stroke %>%
  mutate(
    # Convert "N/A" and "" in bmi to real NA
    bmi = na_if(bmi, "N/A"),
    bmi = na_if(bmi, ""),

    # Convert "Unknown" and "N/A" in smoking_status
    smoking_status = na_if(smoking_status, "Unknown"),
    smoking_status = na_if(smoking_status, "N/A")
  )
```

```r
# ---------------------------------------------
# 4.3 CONVERT BMI BACK TO NUMERIC
# ---------------------------------------------
# After replacing "N/A" with NA, bmi is still character type.
# Convert it to numeric for analysis.
# This might produce a warning:
#   "NAs introduced by coercion" — this is normal.

stroke_clean$bmi <- as.numeric(stroke_clean$bmi)
```

```r
# ---------------------------------------------
# 4.4 CONFIRM MISSING VALUES ARE FIXED
# ---------------------------------------------
# Now we correctly count TRUE NA values in each column.
# You should now see real missing values in bmi and smoking_status.

colSums(is.na(stroke_clean))
```

```r
# -------------------------------------------------------------
# STEP 5: IMPUTE MISSING VALUES
# -------------------------------------------------------------
```

```r
# -----------------------
# 5.1 IMPUTE BMI (NUMERIC)
# -----------------------
# BMI contains numeric values and real NA values.
# Median is used because:
#  • BMI is usually skewed
#  • Median is more robust than mean

# Calculate median BMI without NA values
median_bmi <- median(stroke_clean$bmi, na.rm = TRUE)

# Replace all NA in bmi with the median value
stroke_clean$bmi[is.na(stroke_clean$bmi)] <- median_bmi
```

```r
# --------------------------------------
# 5.2 IMPUTE SMOKING STATUS (CATEGORICAL)
# --------------------------------------
# We use MODE (most frequent category)
# because smoking_status is a factor/character variable.

# Function to compute mode manually
get_mode <- function(x) {
  tab <- table(x)              # Frequency table
  mode_value <- names(tab)[which.max(tab)]  # Most common value
  return(mode_value)
}

# Find the mode of smoking_status
mode_smoking <- get_mode(stroke_clean$smoking_status)

# Replace NA with the mode value
stroke_clean$smoking_status[is.na(stroke_clean$smoking_status)] <- mode_smoking
```

```r
# -------------------------------
# 5.3 CHECK THAT IMPUTATION WORKED
# -------------------------------
colSums(is.na(stroke_clean))
```

```r
# ---------------------------------------------
# STEP 6: Convert character columns to factors
# ---------------------------------------------

stroke <- stroke %>%
  mutate(
    gender = as.factor(gender),
    ever_married = as.factor(ever_married),
    work_type = as.factor(work_type),
    residence_type = as.factor(residence_type),
    smoking_status = as.factor(smoking_status),
    stroke = as.factor(stroke)
  )
```

```r
# -------------------------------------------------------------
# STEP 7: Detect and Remove Outliers in BMI
# -------------------------------------------------------------

# 7.1 Convert BMI to numeric (required for boxplot)
stroke$bmi <- as.numeric(stroke$bmi)

# 7.2 Visualize outliers using a boxplot
boxplot(
  stroke$bmi,
  main = "BMI Outliers",
  ylab = "Body Mass Index"
)

# 7.3 Remove unrealistic BMI values
# In medical datasets, BMI above 60 is extremely unlikely.
stroke <- stroke %>% filter(bmi < 60)

# 7.4 Check summary after removing outliers
summary(stroke$bmi)
```

```r
# ---------------------------------------------
# STEP 8: Feature Engineering
# ---------------------------------------------

stroke <- stroke %>%
  mutate(
    age_group = case_when(
      age < 18 ~ "Child",
      age >= 18 & age < 40 ~ "Young Adult",
      age >= 40 & age < 60 ~ "Middle Age",
      TRUE ~ "Senior"
    ),
    bmi_category = case_when(
      bmi < 18.5 ~ "Underweight",
      bmi >= 18.5 & bmi < 25 ~ "Normal",
      bmi >= 25 & bmi < 30 ~ "Overweight",
      TRUE ~ "Obese"
    )
  )

# Convert new features to factors
stroke$age_group <- as.factor(stroke$age_group)
stroke$bmi_category <- as.factor(stroke$bmi_category)

head(stroke)
```

```r
write_csv(stroke, "stroke_cleaned.csv")
```
